# Widgets
Widgets add interactive input controls to your notebook

- text
- dropdown
- multiselect
- combobox

```python
dbutils.help()
```

```python
dbutils.fs.help()
```

```python
dbutils.widgets.help()
```

```python
dbutils.widgets.text("databaseName", "default")
```

```python
value =dbutils.widgets.get("databaseName")
for i in range(int(value)):
    print(i)
```
