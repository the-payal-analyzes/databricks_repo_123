# All About Notbooks 
**Magic Commands**

Magic commands start with **%** and must be the first line of a cell. They switch the cell's execution context, install packages, run shell commands, or display rich content.

- %md - markdown  , text design for the note book , description 
- %sql - allow you to run sql quries 
- %scala - allow to run scala (not supported in serverless)
- %python - default , allow to run python
- %sh - allow to run shell (linux) commands 
- %fs - databricks file system (dbfs)
