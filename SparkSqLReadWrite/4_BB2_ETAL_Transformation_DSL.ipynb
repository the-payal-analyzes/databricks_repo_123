## select vs selectExpr

```python
stndrd_df6.show(5)

stndrd_df6.select("custid","first_name","last_name").show(5)
stndrd_df6.selectExpr("custid","first_name","last_name").show(5)

# change custid to customer_id

# sql -> select custid as customer_id from tbl

# expr - expression- sql expression 

stndrd_df6.select(col("custid").alias("customer_id"),"first_name","last_name").show(5)
stndrd_df6.selectExpr("custid as customer_id","first_name","last_name").show(5)


stndrd_df6.selectExpr("custid as customer_id","concat(first_name,' ',last_name) as full_name").show(5)

stndrd_df6.selectExpr("custid as customer_id","concat(first_name,' ',last_name) as full_name").show(5)


```
