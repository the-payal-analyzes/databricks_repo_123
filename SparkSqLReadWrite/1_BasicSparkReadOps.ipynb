## How to Read the data from DBFS with various options (CSV)

```python
df=spark.read.csv("dbfs:/Volumes/we48/izdata/data/custs")

df.show()  # action to display the result 

display(df)
```

```python
df=spark.read.csv("dbfs:/Volumes/we48/izdata/data/custs").toDF("custid","firstname","lastname","age","profession")
display(df)
```

```python
%fs head dbfs:/Volumes/we48/izdata/data/custs_header
```

```python
df=spark.read.csv("dbfs:/Volumes/we48/izdata/data/custs_header",header=True)
df.printSchema()
```

```python
df=spark.read.csv("dbfs:/Volumes/we48/izdata/data/custs_header",header=True,inferSchema=True)
display(df)
df.printSchema()
```

```python
schema_str = "custid INT, firstname STRING, lastname STRING, age INT, profession STRING"
df=spark.read.csv("dbfs:/Volumes/we48/izdata/data/custs_header",header=True,inferSchema=False,schema=schema_str)
display(df)
df.printSchema()
```

```python
df.write.saveAsTable("we48.izdata.customer")
```
